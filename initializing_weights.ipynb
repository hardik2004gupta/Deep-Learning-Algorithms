```python
import numpy as np

def initialize_parameters(layer_dims, method="he"):
    """
    layer_dims: list like [n_x, n1, n2, ..., nL]
    method: "he" or "xavier"
    """
    np.random.seed(42)
    parameters = {}
    L = len(layer_dims) - 1

    for l in range(1, L + 1):
        n_prev = layer_dims[l - 1]
        n_curr = layer_dims[l]

        if method == "he":
            parameters[f"W{l}"] = np.random.randn(n_curr, n_prev) * np.sqrt(2 / n_prev)

        elif method == "xavier":
            parameters[f"W{l}"] = np.random.randn(n_curr, n_prev) * np.sqrt(1 / n_prev)

        else:
            raise ValueError("Choose 'he' or 'xavier'")

        parameters[f"b{l}"] = np.zeros((n_curr, 1))

    return parameters
```

```python
layer_dims = [2, 4, 3, 1]

params_he = initialize_parameters(layer_dims, method="he")
params_xavier = initialize_parameters(layer_dims, method="xavier")

print("W1 (He):\n", params_he["W1"])
print("W1 (Xavier):\n", params_xavier["W1"])
```
