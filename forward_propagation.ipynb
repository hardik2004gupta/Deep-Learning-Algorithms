```python
import numpy as np

# Activation functions
def relu(Z):
    return np.maximum(0, Z)

def sigmoid(Z):
    return 1 / (1 + np.exp(-Z))

def tanh(Z):
    return np.tanh(Z)

# Forward propagation
def forward_propagation(X, parameters):
    """
    X: input data (n_features, m)
    parameters: dictionary containing W1, b1, ..., WL, bL
    """
    A = X
    caches = []  # store intermediate values

    L = len(parameters) // 2  # number of layers

    for l in range(1, L):
        W = parameters[f"W{l}"]
        b = parameters[f"b{l}"]

        Z = np.dot(W, A) + b
        A = relu(Z)  # hidden layers use ReLU

        caches.append((A, Z, W, b))

    # Output layer (binary classification → sigmoid)
    WL = parameters[f"W{L}"]
    bL = parameters[f"b{L}"]

    ZL = np.dot(WL, A) + bL
    AL = sigmoid(ZL)

    caches.append((AL, ZL, WL, bL))

    return AL, caches
```

```python
# Example: 3-layer network (2 → 4 → 3 → 1)

np.random.seed(0)

parameters = {
    "W1": np.random.randn(4, 2),
    "b1": np.zeros((4, 1)),
    "W2": np.random.randn(3, 4),
    "b2": np.zeros((3, 1)),
    "W3": np.random.randn(1, 3),
    "b3": np.zeros((1, 1)),
}

# Input (2 features, 5 samples)
X = np.random.randn(2, 5)

AL, caches = forward_propagation(X, parameters)

print("Output:", AL)
```
