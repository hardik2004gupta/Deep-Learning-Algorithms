```python
import numpy as np

# Input (features x samples)
X = np.random.randn(3, 5)

# Layer 1
W1 = np.random.randn(4, 3)
b1 = np.random.randn(4, 1)

# Layer 2
W2 = np.random.randn(2, 4)
b2 = np.random.randn(2, 1)

# Forward through 2 layers
Z1 = np.dot(W1, X) + b1
Z2 = np.dot(W2, Z1) + b2

# Equivalent single layer
W_eq = np.dot(W2, W1)
b_eq = np.dot(W2, b1) + b2

Z_eq = np.dot(W_eq, X) + b_eq

# Compare
print("Difference:", np.abs(Z2 - Z_eq).max())
```

```python
def stack_linear_layers(weights, biases):
    """
    weights: list of W matrices
    biases: list of b vectors
    """
    W_eq = weights[0]
    b_eq = biases[0]

    for i in range(1, len(weights)):
        W_eq = np.dot(weights[i], W_eq)
        b_eq = np.dot(weights[i], b_eq) + biases[i]

    return W_eq, b_eq
```

```python
W3 = np.random.randn(1, 2)
b3 = np.random.randn(1, 1)

# Sequential forward
Z1 = W1 @ X + b1
Z2 = W2 @ Z1 + b2
Z3 = W3 @ Z2 + b3

# Equivalent
W_eq, b_eq = stack_linear_layers([W1, W2, W3], [b1, b2, b3])
Z_eq = W_eq @ X + b_eq

print("Difference:", np.abs(Z3 - Z_eq).max())
```
