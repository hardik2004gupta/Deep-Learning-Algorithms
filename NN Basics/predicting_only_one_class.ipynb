```python
import numpy as np

# XOR-like dataset (non-linearly separable)
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 1, 1, 0])
```

```python
class SimpleModel:
    def __init__(self):
        self.w = np.zeros(2)  # bad initialization (all zeros)
        self.b = 0

    def predict(self, X):
        logits = np.dot(X, self.w) + self.b
        return (logits > 0).astype(int)
```

```python
def debug_single_class_prediction(model, X, y_true=None):
    y_pred = model.predict(X)

    if len(y_pred.shape) > 1:
        y_pred = np.argmax(y_pred, axis=1)

    unique_classes = np.unique(y_pred)
    print("Unique predicted classes:", unique_classes)

    values, counts = np.unique(y_pred, return_counts=True)
    print("\nPrediction distribution:")
    for v, c in zip(values, counts):
        print(f"Class {v}: {c} samples")

    if len(unique_classes) == 1:
        print("\n⚠️ WARNING: Model is predicting ONLY ONE CLASS!")
    else:
        print("\n✅ Model predicts multiple classes")

    if y_true is not None:
        print("\nTrue classes:", np.unique(y_true))
        acc = np.mean(y_pred == y_true)
        print(f"Accuracy: {acc:.4f}")
```

```python
model = SimpleModel()

debug_single_class_prediction(model, X, y)
```
